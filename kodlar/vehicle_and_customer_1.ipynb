### DATAWIG PROBABILITY YAKLAŞIMI

```python
import datawig
```

```python
data = customer

df_train, df_test = datawig.utils.random_split(data)

#Initialize a SimpleImputer model
imputer = datawig.SimpleImputer(
    input_columns=['GENDER_ID','MARITAL_STATUS_ID','BIRTH_DATE','FK_ADDRESS_COMMUNICATION_CITY'], # column(s) containing information about the column we want to impute
    output_column= 'OCCUPATION', # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )

#Fit an imputer model on the train data
imputer.fit(train_df=df_train, num_epochs=10)

#Impute missing values and return original dataframe with predictions
imputed = imputer.predict(df_test)
```

### LİNEAR REGRESSION YAKLAŞIMI

```python
from sklearn.linear_model import LinearRegression

data = customer.copy()

test_data = data[data["OCCUPATION"].isnull()]
data.dropna(inplace=True)

y_train = data["OCCUPATION"]
X_train = data.drop("OCCUPATION", axis=1)
X_test = test_data.drop("OCCUPATION", axis=1)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
df = pd.DataFrame(columns=['occupation_pred'])
for i in y_pred:
    df = df.append({'occupation_pred': round(i,0)}, ignore_index=True)

x = pd.DataFrame(customer.loc[customer.OCCUPATION.isnull() == True]["BASE_CUSTOMER_ID"])
x = x.reset_index()

x = x.drop(['index'],axis=1)
df['BASE_CUSTOMER_ID'] = x

df
```

```python
df.to_csv("occupatiun-linear-regression.csv", index=False)
```
