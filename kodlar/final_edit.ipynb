```python
import datawig
```

```python
data = final

df_train, df_test = datawig.utils.random_split(data)

#Initialize a SimpleImputer model
imputer = datawig.SimpleImputer(
    input_columns=['GENDER_ID','MARITAL_STATUS_ID','BIRTH_DATE','FK_ADDRESS_COMMUNICATION_CITY','FK_RELATION_STATUS_ID','SF_CREATE_DATE','STATUS','REQ_BRAND_CODE','REQ_TOPMODEL_CODE','TRAFFIC_DATE','BRAND_CODE','BASEMODEL_CODE','MOTOR_GAS_TYPE','GEAR_BOX_TYPE'], # column(s) containing information about the column we want to impute
    output_column= 'BIRTH_DATE', # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )

#Fit an imputer model on the train data
imputer.fit(train_df=df_train, num_epochs=10000)

#Impute missing values and return original dataframe with predictions
imputed = imputer.predict(df_test)
```

```python
data = final

df_train, df_test = datawig.utils.random_split(data)

#Initialize a SimpleImputer model
imputer = datawig.SimpleImputer(
    input_columns=['GENDER_ID','MARITAL_STATUS_ID','BIRTH_DATE','FK_ADDRESS_COMMUNICATION_CITY','FK_RELATION_STATUS_ID','SF_CREATE_DATE','STATUS','REQ_BRAND_CODE','REQ_TOPMODEL_CODE','TRAFFIC_DATE','BRAND_CODE','BASEMODEL_CODE','MOTOR_GAS_TYPE','GEAR_BOX_TYPE'], # column(s) containing information about the column we want to impute
    output_column= 'FK_RELATION_STATUS_ID', # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )

#Fit an imputer model on the train data
imputer.fit(train_df=df_train, num_epochs=10000)

#Impute missing values and return original dataframe with predictions
imputed_rel = imputer.predict(df_test)
```

```python
imputed_rel
```

```python
imputed_rel['FK_RELATION_STATUS_ID_imputed'] = round(imputed_rel['FK_RELATION_STATUS_ID_imputed'])
```

```python
for col in imputed_rel.columns:
    print(col)
```

```python
imput = imputed_rel.drop(columns=['Unnamed: 0','VEHICLE_ID','GENDER_ID','MARITAL_STATUS_ID','BIRTH_DATE','FK_ADDRESS_COMMUNICATION_CITY','OCCUPATION','FK_RELATION_STATUS_ID','SALESFILE_ID','GEAR_BOX_TYPE','MOTOR_GAS_TYPE','BASEMODEL_CODE','BRAND_CODE','TRAFFIC_DATE','REQ_TOPMODEL_CODE','REQ_BRAND_CODE','STATUS','SF_CREATE_DATE','CUSTOMER_ID','TOTAL_AMOUNT_GROUP','TOTAL_AMOUNT_TL'])
```

```python
imput = imput.drop(columns=['IS_MAINTENANCE'])
```

```python
imput = imput.rename(columns={"BASE_CUSTOMER_ID": "Id", "FK_RELATION_STATUS_ID_imputed": "Expected"})
```

```python
merge = pd.merge(imput, submit, on = 'Id')
merge.set_index('Id', inplace = True)
```
