```python
import numpy as np
import pandas as pd

sales = pd.read_csv('../yeni_datasets/sales.csv')
servis_bakim = pd.read_csv('../yeni_datasets/servis_bakim.csv')
customer_rel = pd.read_csv('../yeni_datasets/customer_rel.csv')
sifir_arac = pd.read_csv('../yeni_datasets/sifir_arac.csv')
vehicle = pd.read_csv('../yeni_datasets/vehicle-final.csv')
customer = pd.read_csv('../yeni_datasets/customer.csv')
```

```python
final = pd.read_csv('../yeni_datasets/final.csv')
```

```python
# Merge the two dataframes, using _ID column as key
final = pd.merge(customer, customer_rel, on = 'BASE_CUSTOMER_ID')
final.set_index('BASE_CUSTOMER_ID', inplace = True)

# Write it to a new CSV file
final.to_csv('final.csv')
```

```python
# Merge the two dataframes, using _ID column as key
final = pd.merge(final, sales, on = 'CUSTOMER_ID')
final.set_index('CUSTOMER_ID', inplace = True)

# Write it to a new CSV file
final.to_csv('final.csv')
```

```python
# Merge the two dataframes, using _ID column as key
final = pd.merge(final, vehicle, on = 'VEHICLE_ID')
final.set_index('VEHICLE_ID', inplace = True)

# Write it to a new CSV file
final.to_csv('final.csv')
```

```python
# Merge the two dataframes, using _ID column as key
final = pd.merge(final, servis_bakim, on = 'VEHICLE_ID')
final.set_index('VEHICLE_ID', inplace = True)

# Write it to a new CSV file
final.to_csv('final.csv')
```

```python
# Merge the two dataframes, using _ID column as key
final = pd.merge(final, sifir_arac, on = 'VEHICLE_ID')
final.set_index('VEHICLE_ID', inplace = True)

# Write it to a new CSV file
final.to_csv('final.csv')
```
